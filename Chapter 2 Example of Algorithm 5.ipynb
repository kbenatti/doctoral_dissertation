# Fuzzy C-means Equi-balanced

```python
from fuzzy import fuzzy_capacity_constraints, centroides
```

```python
Z = np.ones((n_samples,))
```

```python
LABELS, U = fuzzy_capacity_constraints(X,Z,clusters,2)
```

```python
P = centroides(X, U, 2)
```

```python
for labels in LABELS: 
    mapping_color = def_color(X,labels,clusters)
    for c in range(clusters):
        color = mapping_color[c]
        marker = def_marker(color)
        X0 = X[[labels[i] == c for i in range(len(labels))]]
        plt.plot(X0[:,0], X0[:,1], color+marker)
    for centroide in P:
        plt.plot(centroide[0],centroide[1], 'ko')
    plt.show()
```

```python
plt.hist(U.reshape(n_samples*clusters), bins=100)
plt.show()
```

```python
labels = np.argmax(U, axis=0)
mapping_color = def_color(X,labels,clusters)

for label in range(3):
 
    pertencimento_max = U[label]
    labels = np.ones(len(U[label]))*label
    
    color = mapping_color[label]
    marker = def_marker(color)

    for i in range(100):
        if i == 0:
            indexes = (np.array(pertencimento_max) < (i+1)/100)
        elif i <= 99:
            indexes = (i/100 <= np.array(pertencimento_max)) & (np.array(pertencimento_max) < (i+1)/100)
        else:
            indexes = (i/100 <= np.array(pertencimento_max))
        X_p = X[indexes]
        labels_p = list(np.array(labels)[indexes])
        plt.plot(X_p[:,0], X_p[:,1], color+marker, alpha=0.8*i/100)
    plt.show()
```
