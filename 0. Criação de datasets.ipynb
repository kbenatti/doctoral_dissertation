```python
from utils import def_dataset
import matplotlib.pyplot as plt
```

```python
datasets = {}
count = 1
for n_samples in [120, 240, 360]:
    for clusters in [3,4,5,6]:
        for cluster_std in [2,3]:
            X = def_dataset(n_samples,clusters,cluster_std)
            print('Conjunto',count,'n_samples:',n_samples,'clusters:',clusters,'cluster_std:',cluster_std)
            plt.plot(X[:,0],X[:,1], 'ok')
            plt.show()
            datasets['data_'+str(count)] = {'n_samples':n_samples, 'clusters':clusters, 'cluster_std':cluster_std, 'dataset':X.tolist()}
            count+=1
```

```python
import json

with open('data/datasets.txt', 'w') as json_file:
    json.dump(datasets, json_file)
```

```python
with open ('data/datasets.txt', "r") as json_file:
    datasets = json.loads(json_file.read())
```
